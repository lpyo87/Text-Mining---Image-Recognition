```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("foto_antigua.jpg")#, cv2.COLOR_BGR2RGB) # Se agrega el parámetro de COLOR_BGR2RGB
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### Dimensiones de la imagen

```python
alto = img.shape[0]
ancho = img.shape[1]
```

Siempre se va creando nueva imagen

```python
monocromo = np.zeros((alto, ancho, 3)) #espacio para la imagen resultante
gris_ponderado = np.zeros((alto, ancho, 1)) #seleccionamos 1 canal por que la imagen es en color gris
```

```python
for i in range(0,alto):
    for j in range(0, ancho):
        pixel = img[i,j] #Asignar los pixeles de la imagen.
        
        #Extraemos los canales
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]

        monocromo[i,j] = [0,green,0] #aplicamos transformación monocromo
        gris_ponderado[i,j] = int(0.29*red + 0.59*green + 0.114*blue)
```

### Bajamos imagen a Disco

```python
cv2.imwrite("monocromo.jpg", monocromo)
cv2.imwrite("gris_ponderado.jpg", gris_ponderado)
```
